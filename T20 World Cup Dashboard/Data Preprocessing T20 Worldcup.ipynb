# T20 World Cup Cricket Data Pre Processing

```python
#import the libraries 
import pandas as pd
import numpy as np
import json
```

### Processing Match Results

```python
with open('t20_wc_match_results______20221224_142828.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
```

```python
df_match.head()
```

__Renaming scorecard as  match id to link with other tables.__

```python
df_match.rename({'scorecard':'matchid'}, axis = 1, inplace = True)
```

__Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables__

```python
match_ids_dict = {}

for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['matchid']
    match_ids_dict[key2] = row['matchid']
```

### Process Batting Summary

```python
with open('t20_wc_batting_summary____20221224_143457.json') as f:
    bat = json.load(f)
    all_records = []
    for record in bat:
        all_records.extend(record['battingSummary'])
    
df_batting = pd.DataFrame(all_records)
```

```python
df_batting.info()
```

```python
df_batting.head(10)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x:'not_out' if pd.isnull(x) else 'out')
```

```python
df_batting.head(10)
```

```python
df_batting['out/not_out'].value_counts()
```

```python
df_batting.drop(columns = ['dismissal'], inplace = True)
```

```python
df_batting.head(10)
```

### Removing Weird Characters 

```python
import string

def filter_ascii(s):
    # Create a string of ASCII characters
    ascii_chars = string.ascii_letters + string.digits + string.punctuation + ' '

    # Filter the string to keep only ASCII characters
    s = ''.join([c for c in s if c in ascii_chars])

    return s
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(filter_ascii)
df_batting.head()
```

```python
match_ids_dict['Pakistan Vs England']
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
```

### Process Bowling Summary

```python
with open('t20_wc_bowling_summary___20221224_143547.json') as f:
    ball = json.load(f)
    ball_records = []
    for record in ball:
        ball_records.extend(record['bowlingSummary'])
```

```python
df_bowling  = pd.DataFrame(ball_records)
```

```python
df_bowling
```

```python
df_bowling.shape
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
```

### Process Players Information 

```python
with open('t20_wc_player_info______20221224_143945.json') as f:
    player = json.load(f)
```

```python
df_players= pd.DataFrame(player)
```

```python
df_players
```

Drop the unwanted columns and rows having only NaN Values. 

```python
df_players.drop(columns = ['input', 'warning'], inplace = True)
```

```python
df_players.dropna(subset = ['name'], inplace = True)
```

```python
df_players.reset_index(drop = True, inplace = True)
```

```python
# Remove the non-ascii characters from the name 
df_players['name'] = df_players['name'].apply(filter_ascii)
```

# Saving all the dataframes to csv 

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv', index = False)
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
df_players.to_csv('t20_csv_files/dim_players_no_images.csv', index  = False)
```
