```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
#load train and test files
train = pd.read_csv('train_BRCpofr.csv')
test = pd.read_csv('test_koRSKBP.csv')
```

```python
train_features = train.drop(['cltv'],axis = 'columns')
train_target = train['cltv']
test_features = test.copy()
```

```python
preds = test[['id']]
```

```python
train_features = pd.get_dummies(train_features, drop_first=True)
test_features = pd.get_dummies(test_features, drop_first= True)
train_features.columns = train_features.columns.str.replace('[','').str.replace(']','').str.replace('<','')
test_features.columns = test_features.columns.str.replace('[','').str.replace(']','').str.replace('<','')
```

```python
from xgboost import XGBRFRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor
from catboost import CatBoostRegressor
```

```python
xgbrf = XGBRFRegressor(n_estimators = 500, learning_rate=1)
lgbm = LGBMRegressor(n_estimators=50, learning_rate= 0.1)
gbr = GradientBoostingRegressor(n_estimators= 200, learning_rate = 0.05)
cbr = CatBoostRegressor(n_estimators= 200, learning_rate= 0.05)
```

```python
xgbrf.fit(train_features, train_target)
lgbm.fit(train_features, train_target)
gbr.fit(train_features, train_target)
cbr.fit(train_features, train_target)
```

```python
print(xgbrf.score(train_features, train_target))
print(lgbm.score(train_features, train_target))
print(gbr.score(train_features, train_target))
print(cbr.score(train_features, train_target))
```

```python
preds['xgbrf'] = xgbrf.predict(test_features)
preds['lgbm'] = lgbm.predict(test_features)
preds['gbr'] = gbr.predict(test_features)
preds['cbr'] = cbr.predict(test_features)
```

```python
preds
```

```python
sub = pd.DataFrame()
sub[['id', 'cltv']] = preds[['id', 'xgbrf']]
```

```python
sub.to_csv('xgbrf.csv', index = False)

sub[['id', 'cltv']] = preds[['id', 'lgbm']]
sub.to_csv('lgbm.csv', index = False)

sub[['id','cltv']] = preds[['id','gbr']]
sub.to_csv('gbr.csv',index = False)


sub[['id','cltv']] = preds[['id', 'cbr']]
sub.to_csv('cbr.csv', index = False)
```
